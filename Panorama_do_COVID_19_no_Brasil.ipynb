# Panorama do COVID-19 no Brasil

COVID-19 é uma doença infecciosa causada por recém-descoberto coronavírus.

Transmitido principalmente por meio de gotículas provenientes de tosses ou espirros de pessoas infectadas, a gravidade dos sintomas varia muito de pessoa para pessoa.

Fato é, não se sabe muita coisa a respeito do COVID-19. Estudos estão sendo realizados no mundo todo, porém os resultados ainda não são conclusivos e definitivos.

<center><img alt="Colaboratory logo" width="50%" src="https://raw.githubusercontent.com/carlosfab/dsnp2/master/img/covid19.jpg"></center>

Até o presente momento, observa-se que cerca de 80% dos casos confirmados são assintomáticos e rápidos. A maioria das pessoas que se encaixam nesse grupo, se recupera sem nenhuma sequela.

No entanto, 15% das pessoas terão infecções graves e precisarão de oxigênio. O restante das pessoas, que representam 5%, serão classificadas como infecções muito graves e precisarão de ventilação assistida, por meio de respiradores mecânicos em ambiente hospitalar.

Com o objetivo de elevar a consciência situacional a respeito do COVID-19 no Brasil, irei realizar uma análise sobre os dados públicos da doença.

## Obtenção dos Dados

Os dados utilizados na análise a seguir foram importados do Our World in Data - OWID (Nosso mundo em dados). A OWID é uma organização internacional que disponibiliza dados sobre vários tópicos, seja agricultura, saúde, tecnologia, economia, etc.

A OWID disponibiliza dados relacionados a pandemia de Covid-19 de forma gratuíta e conta com atualizações com curtos períodos de tempo.

** Os dados utilizados nessa análise estão atualizados até o dia `14 de Março de 2021.`

```
# importar as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt


# importar o dataset COVID-19 para um DataaFrame
DATA_PATH = "https://raw.githubusercontent.com/teixeiragthiago/datasets/master/covid-mundo.csv"

df = pd.read_csv(DATA_PATH)
df.date = pd.to_datetime(df.date)
```

### Quais os países com maior numero de casos e mortes?

**Maior número de casos**

```
df.drop(df[df.location.isin(['NaN'])].index, axis=0, inplace=True)

#removendo dados que são referentes a blocos de continentes, estava atrapalhando a listagem dos países
df.drop(df[df.iso_code.isin(['OWID_EUR', 'OWID_NAM', 'OWID_ASI', 'OWID_EUN', 'OWID_SAM'])]
  .index, axis=0, inplace=True)

df.loc[df.date == '2021-03-14'].sort_values(by="total_cases", ascending=False)[1:6]
```

Como a listagem acima nos mostra, podemos verificar que até o dia 14 de março de 2021, os países com maiores números de casos são


*   `Estados Unidos da América`
*   `Brasil`
*   `Índia`
*   `Rússia`
*   `Grã Bretanha`




**Maior número de mortes**

```
df.loc[df.date == '2021-03-14'].sort_values(by="total_deaths", ascending=False)[1:6]
```

Infelizmente, o Brasil se encontra mais uma vez na segunda colocação de uma listagem, mas nesse caso se refere ao total de mortes.

*   `Estados Unidos da América`
*   `Brasil`
*   `México`
*   `Índia`
*   `Grã Bretanha`

Para melhor visualizarmos a quantidade de mortes, vamos analisar o gráfico de barras abaixo:

```
fix, ax = plt.subplots()

top5_total_deaths = df.loc[df.date == '2021-03-14', ['location', 'total_deaths']].sort_values(by="total_deaths", ascending=False)[1:6]
ax.bar('location', 'total_deaths', data=top5_total_deaths)

plt.tight_layout()
```

### Curva de contágio da Covid-19 no mundo


```
df_chart = df.copy()
df_chart.set_index("date", inplace=True)

plot_chart = df_chart.loc[df_chart.location == "World", ['total_cases']]

fig, ax = plt.subplots(figsize=(25,10))

ax.plot('total_cases',data=plot_chart)
ax.set_title("Evolução da Covid-19 no mundo")
ax.set_xlabel("Data")
ax.set_ylabel("Total de casos")

fig.autofmt_xdate()

plt.show()
```

### Análise Exploratória para o Brasil


```
df_brasil = df.loc[df.location == "Brazil"].copy()

df_brasil.head()
```

### Quando foi registrado o primeiro caso no Brasil?

```
df_brasil.loc[df_brasil.new_cases > 0].head(1)
```

Como podemos ver acima, o primeiro caso registrado no brasil ocorreu no dia `26` de `Fevereiro` de `2020`

### Quando foi registrada a primeira morte no Brasil?

```
df_brasil.loc[df_brasil.new_deaths > 0].head(1)
```

Como a pesquisa acima nos mostra, a primeira morte no Brasil foi registrada no dia `17` de `Março` de `2020`

### Qual foi o intervalo de tempo desde o primeiro caso registrado até a primeira morte?

```
from datetime import date

period_interval = df_brasil.date.loc[9634] - df_brasil.date.loc[9614]

print(f"O intervalo desde o primeiro caso confirmado até a primeira morte foi de {period_interval.days} dias.")
```

### Agora, vamos plotar alguns gráficos para visualizarmos melhor a evolução da pandemia no Brasil. Considerando o número de casos e mortes.

```
# Plotar gráficos de linha em escalas linear e logarítmica para os casos totais e mortes totais.
x_axis = list(range(383))
z_axis = df_brasil['total_cases']
z2_axis = df_brasil['total_deaths']

fig, ((ax_one, ax_two), (ax_three, ax_four)) = plt.subplots(nrows=2, ncols=2, figsize=(9, 7))

ax_one.set_title('Casos no Brasil')
ax_one.set_xlabel('Dias')
ax_one.set_ylabel('Casos')
ax_one.plot(x_axis, z_axis)

ax_two.set_title('Casos no Brasil em Escala logarítmica')
ax_two.set_xlabel('Dias')
ax_two.set_ylabel('Casos')
ax_two.set_yscale('log')
ax_two.plot(x_axis, z_axis)

ax_three.set_title('Mortes no Brasil')
ax_three.set_xlabel('Dias')
ax_three.set_ylabel('Mortes')
ax_three.plot(x_axis, z2_axis)

ax_four.set_title('Mortes no Brasil em Escala logarítmica')
ax_four.set_xlabel('Dias')
ax_four.set_ylabel('Mortes')
ax_four.set_yscale('log')
ax_four.plot(x_axis, z2_axis)
plt.tight_layout()
```
